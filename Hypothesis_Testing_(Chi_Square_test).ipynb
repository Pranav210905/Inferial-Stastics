**1. Given the dataset above determine whether there is a relationship between fuel type and body style considering a significance level of 0.5.**

```python
# perform value counts on body-style
data.groupby('body-style')['fuel-type'].value_counts()
```

```python
#pefrom crosstab on body-type and fuel-type
pd.crosstab(data['body-style'], data['fuel-type'], margins=True)
```

- **H0 : Body style and fuel type are independent**
- **H1 : There is a relation between body style and fuel type**

```python
# chi2 test_statistic, p-value, degrees of freedom, expected_frequences
observed=pd.crosstab(data['body-style'],data['fuel-type'])
observed
```

```python
result=chi2_contingency(observed)
result
```

```python
# calculate chi square statistics
chi2_test_stat=result[0]
df=result[2]
expected=result[3]
chi2_test_stat
```

```python
# Critical chi square test statistic
alpha=0.5
chi2_critical=chi2.ppf(1-alpha,df)
chi2_critical
```

```python
if chi2_test_stat > chi2_critical:
    print('Reject Null Hypothesis')
else:
    print('Failed to reject Null Hypothesis')
```
