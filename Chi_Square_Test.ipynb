```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
data=pd.read_csv("/content/drive/MyDrive/Drug.csv")
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.nunique()
```

```python
data.Sex.value_counts()
```

```python
data.Drug.value_counts()
```

```python
data.groupby('Drug')['Sex'].value_counts()
```

```python
pd.crosstab(data.Drug,data.Sex,margins=True)
```

```python
observed=pd.crosstab(data.Drug,data.Sex)
observed
```

```python
result=chi2_contingency(observed)
result
```

```python
chi2_test_stastic=result[0]
degreeoffreedom=result[2]
expected_freq=result[3]
print(chi2_test_stastic)
print(degreeoffreedom)
```

```python
alpha=0.05
chi2_crtitical=chi2.ppf(1-alpha,degreeoffreedom)
chi2_crtitical
```

```python
if(chi2_test_stastic>chi2_crtitical):
  print("We reject the Null Hypothesis")
else:
  print("We don't reject Null Hypothesi")
```
