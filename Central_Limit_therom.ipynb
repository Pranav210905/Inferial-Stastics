```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
random.seed=5
%matplotlib inline

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/drive/MyDrive/Heart Attack Data.csv')
data.head()
```

```python
plt.figure(figsize=(15,6),dpi=200)
sns.distplot(data.age,hist=False)
```

```python
print("No of observations in age",data.age.shape)
```

```python
sample=random.sample(range(1,324),10)
sample
```

```python
data.iloc[sample,0]
```

```python
sample_size=100
no_of_samples=100
sample_means=[]
for i in range(no_of_samples):
  index=random.sample(range(1,324),sample_size)
  sample=data.iloc[index,0]
  sample_means.append(sample.mean())
```

```python
sample_means
```

```python
plt.figure(figsize=(12,5),dpi=200)
sns.distplot(sample_means,hist=False)
```

```python
sample_means_val=np.mean(sample_means)
sample_means_val
```
